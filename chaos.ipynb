```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cobweb(f, x0, N=10, n=2):

    plt.figure(figsize=(20,10))
    yprev = 0
    vals = [x0]
    for i in range(N):
        nx = [x0, f(x0)]
        ny = [f(x0), f(x0)]
        n2x = [x0,x0]
        n2y = [yprev, f(x0)]
        x0 = f(x0)
        yprev = x0
        vals.append(x0)
        plt.plot(nx,ny, "ro--")
        plt.plot(n2x, n2y, "ro--")


    x = np.linspace(min(0,min(vals)),max(vals)*n, 300)
    plt.grid()
    plt.plot(x, f(x))

    plt.plot(x,x)
def orbit(f, x0, n=5):
    res = []
    for i in range(n):
        res.append(x0)
        x0 = f(x0)
    return res

def plot_point(f, r0=0, r1=2):
    x = np.linspace(r0,r1,30000)
    y = f(x)
    plt.plot(x,x)
    plt.plot(x,f(x))
    idx = np.argwhere(np.diff(np.sign(f(x)-x))).flatten()
    plt.plot(x[idx], y[idx], "ro")
    plt.show()
    return x[idx]
```

```python

f = lambda x: x**3-1
print(plot_point(f, -2, 2))
print(orbit(f, 1))
print(orbit(f, 1.4))
print("\n\n=====================================================================\n\n")


f = lambda x: x**2 - 4*x
print(plot_point(f, -2, 10))

print(orbit(f, 1))
print(orbit(f, -2))
print(orbit(f, 3.5))
print(orbit(f, 6))
print("\n\n=====================================================================\n\n")

f = lambda x: np.sin(x)
print(plot_point(f, -2, 3))

print(orbit(f, -1))
print(orbit(f, 1))

print("\n\n=====================================================================\n\n")


f = lambda x: 1/x

print(plot_point(f, -1, 1))

print(orbit(f, -2))
print(orbit(f, -0.7))


print(orbit(f, 1.5))
print(orbit(f, 0.5))



f = lambda x: np.sin(x)/x

print(plot_point(f, -5, 5))

print(orbit(f, 1))
print(orbit(f, -0.7))

```

```python
from tqdm import tqdm
from math import sin,pi
```

```python
def logistic(r, x0):
  while True:
    yield x0
    x0 = r*x0*(1-x0)

def f1(r, x0):
    while True:
        yield x0

        x0 = r*(x0**2)- r*(x0**3)

def f2(r, x0):
    while True:
        yield x0
        x0 = r*sin(pi*x0/2)
```

```python

def bifurcacion(f, x0=0.2, r0 = 0.5, r1 =10):
    scatter = []
    plt.figure(figsize=(20,10))
    rs = np.linspace(r0, r1, 1000)

    for r in tqdm(rs):
        l = f(r, x0)

        x = list(range(1000))
        vals = [next(l) for i in x][900:]

        plt.scatter([r]*len(vals), vals, color="black", s = [1]*len(vals))
    plt.show()
```

```python
bifurcacion(logistic)
```

```python
bifurcacion(f1, x0=0.7, r0=5, r1 = 15)
```

```python
bifurcacion(f2)
```

```python
def gen_orbit(gen, n=40):
    res = []
    for i in range(n):
        res.append(next(gen))

    return res
```
